```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
f_path = './../CelebA-HQ_Dataset/Val/HQ_256x256/'
img_list = glob.glob(f_path + '*.png')
```

```python
for img_name in img_list:
    name = img_name.split('\\')[-1]
    img = cv2.imread(img_name)
    img_128 = cv2.resize(img, (128,128))
    cv2.imwrite('./../CelebA-HQ_Dataset/Val/HQ_128x128/{}'.format(name), img_128)
```
